````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv("Datasets/Iris.csv")

# Display first few rows of the dataset
print("Dataset Overview:")
print(df.head())
````

````python
print("\nFeature Types:")
print(df.dtypes)
````

````python
df=df.drop(columns=['Id'])
````

````python
df.hist(bins=20, figsize=(10, 8))
plt.suptitle("Histogram for Each Feature")
plt.show()
````

````python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, orient='h')
plt.title("Boxplot for Each Feature")
plt.show()

````

````python
sns.pairplot(df, hue='Species')
plt.title('Pairplot of Features')
plt.show()
````

````python
numeric_df = df.select_dtypes(include='number')

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
````

````python
numeric_features = df.select_dtypes(include='number').columns


def detect_outliers(df, feature):
    Q1 = df[feature].quantile(0.25) 
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR 
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
for col in numeric_features:
    outliers = detect_outliers(df, col) 
    print(f"\nOutliers detected in {col}:") 
    if not outliers.empty:
        print(outliers[[col]])
    else:
        print("No outliers detected.")
df_cleaned = df.copy()
for col in numeric_features:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR 
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned = df_cleaned[(df_cleaned[col] >= lower_bound) & (df_cleaned[col] <= upper_bound)]
print("\nShape before handling outliers:", df.shape) 
print("Shape after handling outliers:", df_cleaned.shape)
````

````python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_cleaned, orient='h')
plt.title("Boxplot for Each Feature")
plt.show()

````

---

### 🔹 Why Remove Outliers?

Outliers are data points that deviate significantly from other observations. Here's why we often remove them:

#### ✅ 1. **Improve Model Accuracy**

* Outliers can skew the results of many machine learning models, especially those that rely on distances (e.g., k-NN, SVM) or assume normal distribution (e.g., Linear Regression).
* Removing them can lead to more accurate and stable predictions.

#### ✅ 2. **Better Visualization**

* Outliers can distort plots (like histograms or boxplots), making it harder to see the underlying data distribution.

#### ✅ 3. **Cleaner Statistics**

* Metrics like mean and standard deviation are highly sensitive to outliers. Removing them gives more representative statistics.

#### ✅ 4. **Data Entry Errors**

* Some outliers are just errors (e.g., mistyped values). Removing them improves data quality.

---

### 🔹 What is IQR Technique for Outlier Detection?

**IQR = Interquartile Range**
It's a statistical method used to measure the spread of the middle 50% of your data.

#### 📌 Steps:

1. **Q1 (25th percentile)**: Value below which 25% of the data fall.
2. **Q3 (75th percentile)**: Value below which 75% of the data fall.
3. **IQR = Q3 - Q1**
4. **Outlier bounds**:

   * **Lower bound** = Q1 - 1.5 × IQR
   * **Upper bound** = Q3 + 1.5 × IQR
5. Any data point **< Lower bound or > Upper bound** is considered an **outlier**.

---

### 🔹 Example

Suppose a feature has the following values:

```
[2, 4, 5, 6, 7, 9, 10, 50]
```

* Q1 = 4.5, Q3 = 9.5
* IQR = 9.5 - 4.5 = 5
* Lower Bound = 4.5 - 1.5 × 5 = -3.0
* Upper Bound = 9.5 + 1.5 × 5 = 17.0

**So, 50 is an outlier** because it’s greater than 17.

---

### 🔹 When NOT to Remove Outliers

* If outliers are **genuine and important** (e.g., top salaries, rare diseases).
* In **classification tasks**, sometimes outliers define rare but important cases.
* In **robust models** (like tree-based models), outliers may not hurt much.
