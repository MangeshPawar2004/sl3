````python
# Import necessary libraries
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
df = sns.load_dataset("titanic")

df

````

````python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='sex', y='age', hue='survived')
plt.title('Age Distribution by Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
````

````python
df['age_bin'] = pd.cut(df['age'], bins=[0, 10, 20, 30, 40, 50, 60, 80], labels=['0-10','11-20','21-30','31-40','41-50','51-60','61+'])
````

````python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='age_bin', hue='survived', palette='coolwarm')
plt.title('Survival Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
````

````python
sns.countplot(data=df, x='sex', hue='survived', palette='Accent')
plt.title('Survival by Gender')
plt.show()
````

````python
sns.violinplot(data=df, x='pclass', y='age', hue='survived', split=True, palette='Set2')
plt.title('Age Distribution by Class and Survival')
plt.show()
````

````python
sns.histplot(data=df, x='age', bins=30, kde=True, color='teal')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
````

````python
sns.countplot(data=df, x='embarked', hue='survived', palette='Dark2')
plt.title('Survival by Port of Embarkation')
plt.show()
````

````python
sns.pairplot(df[['age', 'fare', 'pclass', 'survived']], hue='survived', palette='Set1')
plt.suptitle('Pairplot of Age, Fare, Pclass with Survival', y=1.02)
plt.show()
````

## 🧾 PART 1: **Understanding the Titanic Dataset**

The **Titanic dataset** is a classic dataset used in data analysis and machine learning. It contains data of passengers who were aboard the RMS Titanic, which sank in 1912.

### 🎯 **Purpose**:

To analyze factors that influenced the **survival** of passengers.

---

### 🔢 **Key Columns in the Dataset**:

| Column     | Description                                        |
| ---------- | -------------------------------------------------- |
| `survived` | 0 = No, 1 = Yes                                    |
| `pclass`   | Passenger class (1 = upper, 2 = middle, 3 = lower) |
| `sex`      | Gender of passenger                                |
| `age`      | Age in years                                       |
| `sibsp`    | # of siblings/spouses aboard                       |
| `parch`    | # of parents/children aboard                       |
| `fare`     | Ticket fare                                        |
| `embarked` | Port of embarkation (`C`, `Q`, `S`)                |
| `who`      | Man, woman, or child                               |
| `deck`     | Deck location                                      |
| `alone`    | Whether passenger was alone (True/False)           |

## 🧑‍💻 PART 2: **Code Explanation with Theory**

---

### 📌 Step 1: Import Required Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

> **Explanation**:

* `seaborn`: For aesthetic statistical plotting.
* `matplotlib.pyplot`: For chart rendering.
* `pandas`: For structured data manipulation.

---

### 📌 Step 2: Load the Titanic Dataset

```python
df = sns.load_dataset('titanic')
df.head()
```

> **Explanation**: Loads Titanic dataset from Seaborn's built-in datasets. `head()` displays the first 5 rows.

---

### 📌 Step 3: Box Plot for Age vs Gender & Survival

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='sex', y='age', hue='survived', palette='Set3')
plt.title('Age Distribution by Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

> **Theory**:

* A **box plot** shows the distribution of numeric data via **quartiles**.
* **Median**, **Interquartile Range (IQR)**, and **outliers** are shown.
* Useful to compare distributions across categories (e.g., `sex`).

> **What we learn**:

* Median age of female survivors is lower.
* Older males had lower survival rates.

---

### 📌 Step 4: Create Age Bins

```python
df['age_bin'] = pd.cut(df['age'], bins=[0, 10, 20, 30, 40, 50, 60, 80], 
                       labels=['0-10','11-20','21-30','31-40','41-50','51-60','61+'])
```

> **Theory**:

* `pd.cut` creates bins (intervals) for numerical values.
* This helps us convert continuous variables (like age) into categorical age groups for analysis.

---

### 📌 Step 5: Count Plot of Survivors per Age Group

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='age_bin', hue='survived', palette='coolwarm')
plt.title('Survival Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

> **Explanation**:

* A **countplot** counts number of occurrences in each category.
* We compare **survival rate across age groups**.

---

### 📌 Step 6: Survival Count by Gender

```python
sns.countplot(data=df, x='sex', hue='survived', palette='Accent')
plt.title('Survival by Gender')
plt.show()
```

> **Explanation**:

* Females had higher survival due to the **“women and children first”** policy.

---

### 📌 Step 7: Violin Plot of Age by Class

```python
sns.violinplot(data=df, x='pclass', y='age', hue='survived', split=True, palette='Set2')
plt.title('Age Distribution by Class and Survival')
plt.show()
```

> **Theory**:

* **Violin plots** show the **distribution** of the data (like KDE + boxplot).
* Splitting by `survived` helps us compare survivors/non-survivors per class.

---

### 📌 Step 8: Histogram of Age

```python
sns.histplot(data=df, x='age', bins=30, kde=True, color='teal')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

> **Theory**:

* A **histogram** helps visualize the **frequency** of age ranges.
* **KDE (Kernel Density Estimate)** shows smooth probability density.

---

### 📌 Step 9: Count Plot by Embarkation Port

```python
sns.countplot(data=df, x='embarked', hue='survived', palette='Dark2')
plt.title('Survival by Port of Embarkation')
plt.show()
```

> **Theory**:

* Shows survival pattern based on embarkation port (`C`, `Q`, `S`).
* Useful to analyze if location affected survival chances.

---

### 📌 Step 10: Pairplot (Multi-Var Analysis)

```python
sns.pairplot(df[['age', 'fare', 'pclass', 'survived']], hue='survived', palette='Set1')
plt.suptitle('Pairplot of Age, Fare, Pclass with Survival', y=1.02)
plt.show()
```

> **Theory**:

* Pairplot shows **pairwise relationships** in a dataset.
* Good for identifying correlations and trends.

---

## 📊 Final Observations

1. **Women and children** had higher survival rates.
2. **1st class passengers** survived more than 3rd class.
3. **Younger passengers** had better survival rates.
4. **Embarked from Cherbourg (`C`)** = higher survival.
5. **Fare and class** are strong survival indicators.
