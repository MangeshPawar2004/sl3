````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

````python
df = sns.load_dataset('titanic')

# Show first few rows
print(df.shape)
df.head()
````

````python
# Basic info
df.info()

# Summary statistics
df.describe(include='all')


````

````python
# Basic info


# Check for missing values
df.isnull().sum()
````

````python
plt.figure(figsize=(8, 5))
sns.histplot(df['fare'], bins=30, kde=True, color='skyblue')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
````

````python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['fare'], color='orange')
plt.title("Boxplot of Passenger Fare")
plt.xlabel("Fare")
plt.show()
````

````python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='sex', hue='survived', palette='Set2')
plt.title("Survival Count by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
````

````python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='class', hue='survived', palette='Set1')
plt.title("Survival by Passenger Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.show()
````

````python
plt.figure(figsize=(8, 5))
sns.histplot(data=df, x='age', hue='survived', bins=30, kde=True, palette='husl')
plt.title("Age Distribution by Survival")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
````

````python
sns.scatterplot(x="age", y="fare", hue="survived", style="sex", data=df, palette="deep")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatter Plot of Age vs Fare (Colored by Survival Status)")
plt.show()

````

````python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
````

````python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='class', y='age', hue='sex', split=True, palette='Set3')
plt.title("Age Distribution by Class and Gender")
plt.show()
````

Titanic dataset reveals social patterns  survival highly influenced by class and gender
